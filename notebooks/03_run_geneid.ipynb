```python
species = ! ls ../results/geneid_predictions/

with open('../job_commands/SAR_pred.txt', 'w') as out:
    
    for sp in species:
        ref = ! realpath ../results/geneid_predictions/$sp/*fna
        param = ! realpath ../results/geneid_predictions/$sp/*param
        models = f'/users/rg/fzanarello/projects/geneid-bg/results/geneid_predictions/{sp}/{sp}.gff3'
    
        command = f'time geneid -3P {param[0]} {ref[0]} > {models}'
        out.write(command + '\n')
```
